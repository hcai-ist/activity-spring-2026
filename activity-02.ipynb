````python
import getpass
import os

import instructor
from openai import OpenAI
from pydantic import BaseModel
````

````python
# You should NEVER share your API keys publicly.

# In this case, we are going to save it to a variable


_API_KEY = getpass.getpass("Your API key: \n")
````

````python
client = instructor.from_provider(
    "openai/gpt-4o-mini", mode=instructor.Mode.RESPONSES_TOOLS, api_key=_API_KEY
)
````

````python
class Response(BaseModel):
    message: str
````

## You can give instructions about how to perform a task

* [Message roles and instruction following](https://platform.openai.com/docs/guides/text?lang=python#message-roles-and-instruction-following)

````python
r = client.responses.create(
    input="Write a haiku about IST 597 (HCAI): a class focusing on human-centered AI",
    instructions="In Spanish",
    response_model=Response,
)

print(r.message)
````

## Roles and authority

* [Instructions and levels of authority](https://model-spec.openai.com/2025-02-12.html#:~:text=caution.-,Instructions%20and%20levels%20of%20authority,-While)
* Platform > Developer > User > Guideline

````python
roles = [
    {"role": "developer", "content": "Talk like a pirate."},
    {"role": "user", "content": "Describe the weather today in State College, PA"},
]

r = client.responses.create(input=roles, response_model=Response)
print(r.message)
````
